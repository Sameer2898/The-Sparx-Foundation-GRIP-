```python
# Import the important libraries
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
iris= sns.load_dataset('iris')
iris.keys()
```

```python
# Print the description of the iris data
print(f'Information of the dataset:-\n\n{iris.info()}')
```

```python
# Print first and last five rows of the data
print(f'Head of the dataset:-\n\n{iris.head()}')
print(f'\n\nTail of the dataset:-\n\n{iris.tail()}')
```

```python
# Preparing the training and testing set
x = iris.iloc[:, :-1]
y = iris.iloc[:, -1]

print(f'Features Values:-\n\n{x}')
print(f'\n\nTarget Values:-\n\n{y}')
```

```python
# Plot the feature and species
plot_x = iris.loc[:, 'sepal_length']
plot_y = iris.loc[:, 'species']
plt.scatter(plot_x, plot_y, color='blue', label='sepal_length')

plot_x = iris.loc[:, 'sepal_width']
plot_y = iris.loc[:, 'species']
plt.scatter(plot_x, plot_y, color='green', label='sepal_width')

plot_x = iris.loc[:, 'petal_length']
plot_y = iris.loc[:, 'species']
plt.scatter(plot_x, plot_y, color='red', label='petal_length')

plot_x = iris.loc[:, 'petal_width']
plot_y = iris.loc[:, 'species']
plt.scatter(plot_x, plot_y, color='yellow', label='petal_width')

plt.xlabel('Features')
plt.ylabel('Species')
plt.legend(loc=4, prop={'size':8})
plt.show()
```

```python
# Split the data in to training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f'X Train:-\n{x_train}')
print(f'\nX Test:-\n{x_test}')
print(f'\nY Train:-\n{y_train}')
print(f'\nY Test:-\n{y_test}')
```

```python
# Train the model
# model = LogisticRegression()
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# Test the model
predictions = model.predict(x_test)
print(f'Predictions:-\n{predictions}')
print(f'Testing Data:-\n{y_test}')
```

```python
# Predict the value on the basis of the given data
ans = model.predict([[6, 3, 4, 2]])
print(f'The type of the flower according to the given value:-\n{ans}')
```

```python
model_all = DecisionTreeClassifier().fit(x, y)
plt.figure(figsize=(20, 10))
plot_tree(model_all, filled=True, rounded=True)
plt.show()
```

```python
# Check precision, recall and f1-score
print(f'Classification Report:-\n\n{classification_report(y_test, predictions)}')
```

```python
# Accuracy score
print(f'Accuracy Score:-\n{accuracy_score(y_test, predictions)}')
```
